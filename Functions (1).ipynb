Q1. What is the difference between a function and a method in Python?

 In Python, the primary difference is that a function is a standalone block of code called by its name, while a method is a function that is explicitly bound to an object or a class.

 While both are reusable blocks of code defined using the def keyword, they differ fundamentally in how they are defined, called, and how they interact with data.

# Understanding Functions
 A function is a global, independent entity. It takes inputs, executes logic, and optionally returns a value. Python comes with built-in functions
 print(), len(), or sum(), and you can also build custom user-defined functions.Code Example:python# Defining a standalone function

    def calculate_area(width, height):
    return width * height

    # Calling the function independently
    result = calculate_area(5, 10)
    print(result)  # Output: 50
#Understanding Methods
A method is a function nested inside a class. It represents the behavior or action an object can take. Because it is bound to an object instance, its first parameter is conventionally named self, which Python automatically passes behind the scenes when you invoke the method.Code Example:python# Defining a class with a method

     class Rectangle:
       def __init__(self, width, height):
          self.width = width   # Object attribute
           self.height = height # Object attribute

    # This is a method because it belongs to the class
    def get_area(self):
        return self.width * self.height

    # Creating an object (instance) of the class
      my_rectangle = Rectangle(5, 10)
    # Calling the method on the object using dot notation
    # 'my_rectangle' is implicitly passed into 'self'
     print(my_rectangle.get_area())  # Output: 50

 # Deep Dive: Advanced Method Types
Not all methods behave identically in Python. Depending on the decorators you use, methods can be split into three categories:

**Instance Methods:** The standard method type (like get_area above). They take self as the first argument and can freely modify individual object states

**Class Methods:** Decorated with @classmethod, they take the class itself (cls) as the first argument instead of an instance. They can modify the state of the class as a whole rather than a specific object.

**Static Methods:** Decorated with @staticmethod, they are packed inside a class for logical grouping but do not take a self or cls argument. They act exactly like regular functions but live inside the class namespace

Q2. Explain the concept of function arguments and parameters in Python.

Parameters are placeholders defined in a function's signature, while arguments are the actual values passed to those placeholders when the function is invoked. While people often use these terms interchangeably, keeping them distinct helps you master how data moves through your Python code.

    # 'name' is the parameter (the variable placeholder)
    def greet(name):
    print(f"Hello, {name}!")

    # "Alice" is the argument (the actual data passed)
     greet("Alice")

#Types of Arguments in Python
Python offers highly flexible ways to map your arguments to your defined parameters during a function call:

**1. Positional Arguments**  These are matched directly to the parameters based on the order you pass them.
* Order matters.
* Swapping the data alters how the function processes it.

      def describe_pet(animal, name):
      print(f"I have a {animal} named {name}.")

      # Correct order
      describe_pet("dog", "Buddy")  # Output: I have a dog named Buddy.

**2. Keyword Arguments** You pass these by explicitly naming the parameter alongside the value using an equals sign (=).
* Order does not matter.
* Greatly improves code readability.

       # The order is flipped, but the output remains correct
      describe_pet(name="Buddy", animal="dog")

**3. Default Arguments**  You can assign default values to parameters when defining the function.
* If an argument is omitted during the call, Python uses this fallback value.Optional to provide during a call
* Non-default parameters must always be placed before default ones in your code definition.      

      def greet_user(name, message="Welcome"):
      print(f"{message}, {name}!")

      greet_user("Elena")             # Output: Welcome, Elena!
      greet_user("Elena", "Good morning") # Output: Good morning, Elena!

 **4. Arbitrary Arguments (*args and **kwargs)**
When you do not know beforehand how many inputs your function will receive, use these special syntax options:

*  **args (Arbitrary Positional Arguments): Packs an unspecified number of positional inputs into a single tuple.
* **kwargs (Arbitrary Keyword Arguments): Packs an unspecified number of keyword inputs into a single dictionary.

      def Make_pizza(size, *toppings):
      print(f"Making a {size} pizza with toppings: {toppings}")

      # "pepperoni" and "olives" are packed into a tuple
      make_pizza("large", "pepperoni", "olives")

Q5. What are iterators in Python and how do they differ from iterables

An **iterator** in Python is an agent object that traverses a collection and fetches elements one by one, while an **iterable** is the data container itself that can be looped over.

#Understanding Iterables
An iterable is any Python object capable of returning its members one at a time, making it compatible with a for loop. Under the hood, an iterable defines the

1.__iter__() method. When Python needs to loop through an iterable, it passes the object to the iter() function to generate a brand new iterator.

     my_list = [1, 2, 3]  # This is an iterable

     # You can loop over it multiple times
      for item in my_list:
      print(item)
#Understanding Iterators
An iterator is the actual object that performs the iteration. It implements the Iterator Protocol, which requires two magic methods:
* __iter__(): Returns the iterator object itself.
* __next__(): Returns the next item from the sequence.    

You fetch items from an iterator using the next() function. Once the iterator runs out of data, any further calls to next() will raise a StopIteration exception. Crucially, iterators are exhaustible; once you traverse them, they become empty.

      my_list = [1, 2, 3]
      my_iterator = iter(my_list)  # This creates an iterator

      print(next(my_iterator))  # Outputs: 1
      print(next(my_iterator))  # Outputs: 2
      print(next(my_iterator))  # Outputs: 3

     # This final call will raise a StopIteration exception
     # print(next(my_iterator))
#The Relationship
A vital rule in Python is that every iterator is also an iterable, but not every iterable is an iterator. Because iterators implement __iter__() (which just returns self), you can pass an iterator directly into a for loop. However, standard collections like lists do not implement __next__(), meaning they are not iterators on their own.
#How the for Loop Ties Them Together
When you write a standard loop like for x in my_list:, Python executes the following workflow automatically behind the scenes:

1. Calls iter(my_list) to obtain an iterator object.
2. Calls next(iterator) repeatedly to fetch each element.
3. Catches the StopIteration exception silently to exit the loop cleanly without crashing.

Q6. Explain the concept of generators in Python and how they are defined

A Python generator is a special type of function that returns a lazy iterator object, producing a sequence of values one at a time on demand instead of storing the entire dataset in memory. This makes generators highly memory-efficient and ideal for working with massive datasets, real-time data streams, or infinite loops.
# Key Concepts
**Lazy Evaluation:** Values are calculated only when you explicitly request them.                
**State Retention:** When a generator pauses, it saves its local variable states and execution point. It resumes exactly where it left off.                    
**Automatic Iterator Creation:** Python automatically implements the __iter__() and __next__() methods behind the scenes, saving you from writing boilerplates.   
#How Generators are Defined
There are two primary ways to create a generator in Python:
#1. Generator Functions (yield keyword)
You define a generator function exactly like a regular function using def. However, instead of using return to terminate the function and send back a final result, you use yield to produce a value and pause execution.

    def count_up_to(maximum):
       count = 1
       while count <= maximum:
          yield count  # Execution pauses here, saving the local state
          count += 1   # Execution resumes here on the next call
#How to use it:
Calling the function does not run the code; it returns a generator object. You extract values using a for loop or the next() function.

      counter = count_up_to(3)

      print(next(counter))  # Output: 1
      print(next(counter))  # Output: 2
      print(next(counter))  # Output: 3
       # Calling next(counter) again would raise a StopIteration exception

#2. Generator Expressions
For simpler tasks, you can use a generator expression. They look identical to list comprehensions but use parentheses () instead of square brackets [].

       # List comprehension (Loads all 1 million squares into RAM immediately)
       square_list = [x**2 for x in range(1000000)]

       # Generator expression (Produces items one-by-one as requested)
       square_generator = (x**2 for x in range(1000000))

Q7. What are the advantages of using generators over regular functions?

Generators provide major advantages over regular functions by enabling lazy evaluation and better memory efficiency. Instead of computing and returning an entire dataset at once, generators yield values on-the-fly, allowing you to easily handle massive streams of data, maintain state across iterations, and write cleaner, highly optimized code.
#The primary advantages include:

*  **Memory Efficiency:** Generators compute values one at a time and do not store the entire dataset in memory. This is ideal for processing massive files or large data streams where regular functions would cause a memory overload.

*  **Lazy Evaluation:** Values are generated only when requested. This saves processing time and resources, as the program only evaluates what it immediately needs rather than computing unused data.

*  **Infinite Sequences:** Because they evaluate lazily, generators can represent infinite data streams (like continuous sensor readings or an unending sequence of numbers) that would crash a regular function attempting to return a complete list.

*  **Maintaining State:** Unlike regular functions that start from scratch with local variables reset on every call, generators "pause" and "resume" execution. They remember their exact state, making it easier to write iterative algorithms.

*  **Improved Performance:** By avoiding the creation and garbage collection of large lists, generators reduce overall overhead and speed up your program.

Q9. Explain the purpose and usage of the `map()` function in Python.

The purpose of Python's built-in map() function is to apply a specific function to every element of an iterable (such as a list, tuple, or string) without using an explicit loop. It transforms data cleanly and returns a lazy iterator (a map object), meaning it processes elements on demand to save memory.

     map(function, iterable, ...)

**function:** The operation or transformation logic to apply.
**iterable:** One or more collections (lists, tuples, etc.) containing the data you want to transform.

#Key Usage Patterns

1. Using with Built-In FunctionsYou can pass existing Python functions directly into map(). Remember to pass the function name without parentheses.

    # Convert a list of numeric strings into actual integers
     string_numbers = ["1", "2", "3", "4"]
     integers = map(int, string_numbers)

     # Convert the lazy iterator to a list to see the results
     print(list(integers))  
     # Output: [1, 2, 3, 4]


2. Using with Lambda (Anonymous) FunctionsFor simple, custom calculations, pairing map() with a lambda function keeps your code concise.
    

       # Square each number in a list
       numbers = [1, 2, 3, 4]
       squared = list(map(lambda x: x ** 2, numbers))

        print(squared)  
       #  Output: [1, 4, 9, 16]


3. Using with Multiple IterablesIf the function you provide expects multiple arguments, you can pass multiple iterables in parallel. The execution automatically stops as soon as the shortest iterable runs out of elements.

       # Add elements of two lists together position-by-position
       list_a = [1, 2, 3]
       list_b = [10, 20, 30]

       sums = list(map(lambda x, y: x + y, list_a, list_b))

        print(sums)  
        # Output: [11, 22, 33]
#Important Things to Remember

**Lazy Evaluation:** map() does not calculate results all at once. It waits until you iterate over it (e.g., using a for loop or wrapping it in list()). This makes it highly efficient for massive datasets.

**Single-Use Iterator:** Because the result is an iterator, it gets exhausted after you read it once. If you need to use the data multiple times, save it as a concrete list

**Alternative:** In modern Python, list comprehensions (e.g., [x**2 for x in numbers]) are often preferred over map() because many developers find them cleaner and easier to read.

 Q10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

The core difference is that map() transforms every element in an iterable, filter() selects specific elements based on a True/False condition, and reduce() aggregates all elements into a single, combined value
#Execution with your list: [47, 11, 42, 13]
 1. **map**(function, iterable)Applies a specific math operation or function to every single element.
 * Example Action: Multiply every number in your list by
 *  Code

          list(map(lambda x: x * 2, [47, 11, 42, 13]))
          Output: [94, 22, 84, 26]

2. **filter**(function, iterable)Evaluates each element against a condition that returns True or False, keeping only the items that evaluate to True.

*  Example Action: Keep only the even numbers from your list.
*  Code:

     list(filter(lambda x: x % 2 == 0, [47, 11, 42, 13]))
    Output: [42] (47, 11, and 13 are odd numbers, so they are discarded)
 .3. **reduce**(function, iterable)Takes the first two elements, applies a binary operation, takes that result alongside the third element, and repeats this rolling computation until one final output remains.

 *  Example Action: Calculate the total sum of all items in your list.
 * Code:

     from functools import reduce
     reduce(lambda x, y: x + y, [47, 11, 42, 13])
     Output: 113 (Calculated step-by-step as: 47 + 11 = 58 → 58 + 42 = 100 → 100 + 13 = 113)

The internal mechanism of the reduce() function processes the list [47, 11, 42, 13] step-by-step by accumulating a single running total using a two-argument function (like an addition operation)

#🪵 Visual Tree Breakdown
If you were drawing this on paper, the reduction forms a left-leaning evaluation tree:text

       (Final Result: 113)
             /      \
          100        13
         /   \
       58     42
      /  \
    47    11
#⚙️ Step-by-Step Logic

*  **First Pass:**
 The function pulls the first two elements from the list (47 and 11) and adds them together to get 58.

 *  **Second Pass:** The previous result (58) becomes the new left argument, and the next element in the list (42) is pulled to calculate 58 + 42 = 100.

 * **Third Pass:** The updated running total (100) is added to the final element (13) to reach 113.

 *  **Return:** The list is now fully consumed, and reduce() returns the final total of 113.

```python
def sum_even_numbers(numbers):
    total = 0
    for num in numbers:
        if num % 2 == 0:
            total += num
    return total

# Example usag
nums = [1, 2, 3, 4, 5, 6, 7, 8]
print("Sum of even numbers:", sum_even_numbers(nums))
```

```python
def is_prime(num):
    if num < 2 or num > 200:
        return "Please enter a number between 1 and 200."

    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            return False
    return True

# Example usage
number = int(input("Enter a number (1-200): "))

if is_prime(number) == True:
    print(number, "is a prime number.")
elif is_prime(number) == False:
    print(number, "is not a prime number.")
else:
    print(is_prime(number))
```

```python
class FibonacciIterator:
    """An iterator that generates the Fibonacci sequence for a set number of terms."""

    def __init__(self, max_terms):
        self.max_terms = max_terms
        self.count = 0
        self.a = 0
        self.b = 1

    def __iter__(self):

        return self

    def __next__(self):
        if self.count >= self.max_terms:
            raise StopIteration

        current_term = self.a

        self.a, self.b = self.b, self.a + self.b

        self.count += 1

        return current_term
fib_sequence = FibonacciIterator(8)

print("Using a loop:")
for num in fib_sequence:
    print(num, end=" ")
print("\n")
print("Converted directly to a list:")
fib_list = list(FibonacciIterator(5))
print(fib_list)
```
